Task 1: Basic Line ***Plot***

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.show()
```

Task 2: Scatter ***Plot***

```python
# Data points
x_scatter = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y_scatter = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, s=100, c='blue', alpha=0.7, edgecolors='black')
plt.title('Scatter Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.show()
```

Task 3: Bar ***Plot***

```python
# Data points
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Task 4: ***Histogram***

```python
import numpy as np

# Generate random numbers
data = np.random.randn(1000)

# Create a histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='cyan', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Task 5: Pie *Chart*

```python
# Data points
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]

# Create a pie chart
plt.figure(figsize=(8, 5))
explode = (0.1, 0, 0, 0)  # Explode the first slice (Python)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, startangle=140)
plt.title('Pie Chart of Programming Languages')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```
